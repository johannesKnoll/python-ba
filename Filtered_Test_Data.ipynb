```python
def filter_testdata(array, test_array):
    
    include_item = set()
    result = []
    expressions_remove = []

    for item in array:
        if item not in include_item:
            include_item.add(item)
            result.append(item)
    
    
    range_array = len(test_array)
    
    for index in range(0, range_array):
        for index_array in range(0, len(result)):
            if test_array[index] == result[index_array]:
                expressions_remove.append(result[index_array])
                
    for index in range(0, len(expressions_remove)):
        test_array.remove(expressions_remove[index])
                
    return test_array


def cut_array(array_to_cut, other_array):
    cut_length = len(array_to_cut) - len(other_array)
    
    for index in range(0, cut_length):
        array_to_cut.pop(0)
        
    
    return array_to_cut
```
